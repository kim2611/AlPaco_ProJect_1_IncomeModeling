```python
from sklearn.preprocessing import LabelEncoder

le_hot = LabelEncoder()
le_rain = LabelEncoder()
le_cloud = LabelEncoder()
le_mois = LabelEncoder()
le_direction = LabelEncoder()
le_winds = LabelEncoder()
le_dust = LabelEncoder()
le_disc = LabelEncoder()
le_derby = LabelEncoder()
le_holiday = LabelEncoder()

df['폭염여부'] = le_hot.fit_transform(df['폭염여부'])
df['강수여부'] = le_rain.fit_transform(df['강수여부'])
df['구름여부'] = le_cloud.fit_transform(df['구름여부'])
df['습도여부'] = le_mois.fit_transform(df['습도여부'])
df['풍향'] = le_direction.fit_transform(df['풍향'])
df['바람세기'] = le_winds.fit_transform(df['바람세기'])
df['미세먼지 수준'] = le_dust.fit_transform(df['미세먼지 수준'])
df['할인여부'] = le_disc.fit_transform(df['할인여부'])
df['더비여부'] = le_derby.fit_transform(df['더비여부'])
df['휴일 여부'] = le_holiday.fit_transform(df['휴일 여부'])
```

```python
X = df[['폭염여부','강수여부', 'X구름여부', '습도여부', '풍향', '바람세기', '미세먼지 수준','더비여부', '휴일 여부']]
y = df['할인여부']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size =0.1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(X_train_scaled, y_train)
```

```python
clf.score(X_train_scaled, y_train)
```

```python
clf.score(X_test_scaled, y_test)
```

```python
y_pred = clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```
