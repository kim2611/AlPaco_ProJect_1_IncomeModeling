```python
import pandas as pd
main = pd.read_csv('/content/K리그매치데이_기상 데이터+미세먼지농도 (3).csv')
```

```python
main.head()
```

```python
#데이트 타임
main['일시'] = pd.to_datetime(main['일시'])
```

```python
main.columns
```

```python
#데이터 추출
main_0 = main[['일시','관람인원', '수입금(원)', '사용일수(일)', '최고기온(°C)', '1시간 최다강수량(mm)', '최다풍향(16방위)', '최대 풍속(m/s)', '평균 전운량(1/10)','평균 상대습도(%)', '1일평균 미세먼지농도(㎍/㎥)']]
```

```python
#강수량 함수 정의
def rain(x):
    if 1 <= x < 3:
        return "약한 비"
    elif 3 <= x < 15:
        return "보통"
    elif 15 <= x < 30:
        return "강한 비"
    elif 30 <= x:
        return "매우 강한 비"
    else:
        return "맑음"
```

```python
# 풍향 함수 정의
def direction(x):
    if 22.5 < x <= 67.5:
        return "북동"
    elif 67.5 < x <= 112.5:
        return "동"
    elif 122.5 < x <= 157.5:
        return "남동"
    elif 157.5 < x <= 202.5:
        return "남"
    elif 202.5 < x <= 247.5:
        return "남서"
    elif 247.5 < x <= 292.5:
        return "서"
    elif 292.5 < x <= 337.5:
        return "북서"
    else:
        return "북"
```

```python
# 전운량 함수 정의
def cloud(x):
    if 8.5< x:
        return '흐림'
    elif 5.5 < x <= 8.5:
        return '구름 많음'
    else:
        return '맑음'
```

```python
# 풍속 함수 정의
def windspeed(x):
    if 4<= x < 9:
        return '약풍'
    elif 9<= x:
        return '강풍'
    else:
        return '무풍'
```

```python
# 폭염 함수 정의
def tem(x):
    if 33< x:
        return "폭염"
    else:
        return '일반'
```

```python
# 미세먼저 수준 정의
def dust(x):
    if 30 < x <=80:
        return '보통'
    elif 80 < x <= 150:
        return '나쁨'
    elif 150< x:
        return '매우 나쁨'
    else:
        return '좋음'
```

```python
# 습도 수준 정의
def mois(x):
    if x <40 :
        return '건조'
    elif x > 60 :
        return '다습'
    else:
        return '보통'
```

```python
#데이터 전처리 시작
main_0.info()
```

```python
main_0['폭염여부'] = main_0['최고기온(°C)'].apply(tem)
```

```python
main_0['강수여부'] = main_0['1시간 최다강수량(mm)'].apply(rain)
```

```python
main_0['구름여부'] = main_0['평균 전운량(1/10)'].apply(cloud)
```

```python
main_0['습도여부'] = main_0['평균 상대습도(%)'].apply(mois)
```

```python
main_0['풍향'] = main_0['최다풍향(16방위)'].apply(direction)
```

```python
main_0['바람세기'] = main_0['최대 풍속(m/s)'].apply(windspeed)
```

```python
main_0['미세먼지 수준'] = main_0['1일평균 미세먼지농도(㎍/㎥)'].apply(dust)
```

```python
main_0.columns
```

```python
main_1 = main_0[['일시','관람인원','수입금(원)','사용일수(일)','폭염여부', '강수여부', '구름여부', '습도여부', '풍향', '바람세기', '미세먼지 수준']]
```

```python
main_1.to_csv('1차 전처리 완료.csv', index=False, encoding = 'cp949') #1차 저장
```
