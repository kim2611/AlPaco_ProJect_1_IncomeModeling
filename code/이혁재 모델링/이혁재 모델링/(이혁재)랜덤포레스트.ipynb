```python
import pandas as pd
df = pd.read_csv('/content/1차 전처리 완료_최종.csv', encoding = 'cp949')
```

```python
from sklearn.preprocessing import LabelEncoder

le_hot = LabelEncoder()
le_rain = LabelEncoder()
le_cloud = LabelEncoder()
le_mois = LabelEncoder()
le_direction = LabelEncoder()
le_winds = LabelEncoder()
le_dust = LabelEncoder()
le_disc = LabelEncoder()


df['폭염여부'] = le_hot.fit_transform(df['폭염여부'])
df['강수여부'] = le_rain.fit_transform(df['강수여부'])
df['구름여부'] = le_cloud.fit_transform(df['구름여부'])
df['습도여부'] = le_mois.fit_transform(df['습도여부'])
df['풍향'] = le_direction.fit_transform(df['풍향'])
df['바람세기'] = le_winds.fit_transform(df['바람세기'])
df['미세먼지 수준'] = le_dust.fit_transform(df['미세먼지 수준'])
df['할인여부'] = le_disc.fit_transform(df['할인여부'])
```

```python
X = df[['폭염여부','강수여부', '구름여부', '습도여부', '풍향', '바람세기', '미세먼지 수준']]
y = df['할인여부']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size =0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 모델링(지도학습, 분류)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```
