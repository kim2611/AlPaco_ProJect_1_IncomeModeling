```python
import pandas as pd
df = pd.read_csv('/content/1차 전처리 완료_최종2 더비 휴일 포함.csv', encoding = 'cp949')
```

```python
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf

import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
```

```python
gr = df.groupby('더비여부')['관람인원'].mean()
```

```python
df = pd.DataFrame(gr)
df = df.reset_index()
pd.options.display.float_format = '{:.0f}'.format
```

```python
df
```

```python
plt.bar(df['더비여부'], df['관람인원'], color = ['skyblue', 'orange'])
plt.axhline(y=16040, color='gray', linestyle='--')
plt.text(x=0, y=16040, s='16040', ha='center', va='bottom', color='orange')
# plt.yticks(range(0, 45000001, 10000000), [f"{i//10000000}천만원" for i in range(0, 45000001, 10000000)])
# x축 레이블 설정
plt.xlabel('더비여부')

# y축 레이블 설정
plt.ylabel('평균관중')

# 그래프 타이틀 설정
plt.title('더비여부에 따른 관람 평균 관중')


# 그래프 보여주기
plt.show()
```
