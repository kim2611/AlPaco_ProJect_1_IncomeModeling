```python
from sklearn.preprocessing import LabelEncoder

df[['폭염여부', '강수여부', '구름여부', '습도여부', '풍향', '바람세기', '미세먼지 수준', '할인여부']] = df[['폭염여부', '강수여부', '구름여부', '습도여부', '풍향', '바람세기', '미세먼지 수준', '할인여부']].apply(LabelEncoder().fit_transform)
```

```python
df
```

```python
df.groupby('할인여부')['일시'].count()
```

```python
X = df[['폭염여부','강수여부', '구름여부', '습도여부', '풍향', '바람세기', '미세먼지 수준']]
y = df['할인여부']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3, random_state=0)
print(X_train.shape) 
print(X_test.shape) 
print(y_train.shape) 
print(y_test.shape) 
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

pre = DecisionTreeClassifier(random_state=0)
pre.fit(X_train_scaled, y_train)
```

```python
pre.score(X_train_scaled, y_train)
```

```python
pre.score(X_test_scaled, y_test)
```

```python
y_pred = pre.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```
